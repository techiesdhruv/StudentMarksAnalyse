```python
df['Performance'] = df[['MathScore', 'ReadingScore', 'WritingScore']].mean(axis=1)
df['PerfLevel'] = pd.cut(df['Performance'], bins=[0, 50, 70, 90, 100],
                         labels=['Poor', 'Average', 'Good', 'Excellent'])
sns.countplot(data=df, x='PerfLevel')
plt.show()
```

```python
# Parentseducation effect on student marks
```

```python
gb=df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb)
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(gb,annot=True)
plt.title('ParentsEdu vs Student marks avg')
plt.show()
```

```python
#From the above chart that we have concluded that the education of parents affect the student Marks.
#the higher the parents's educatin the higher chance of getting good marks by students.
```

```python
gb1=df.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb1)
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(gb1,annot=True)
plt.title('ParentsMaritalStatus vs Student marks avg')
plt.show()
```

```python
#from the above chart we have concluded that parentmaritalstatus doesnot affect the student marks
```
